```python
import pandas as pd
df = pd.read_csv("/Users/MacBook/Desktop/project3.txt", sep="\t", header=0, index_col=False) #load in data

df['Avg'] = df.iloc[:, 1:5].mean(axis=1) #find the avg of values and creat column called "Avg"
df.sort_values(by='GeneSymbol') #sort alphabetically by gene name
alphabet_df = df.iloc[:,[0,5]] #sorted df
print("***PART A***\n")
print(alphabet_df.to_string(index=False, header=False)) #sorted alphabetically with paired averages
print("\n")
sorted_df = alphabet_df.sort_values(by='Avg') #sort by avg low to high
sorted_df = sorted_df.iloc[:,0]
print(sorted_df.to_string(index=False, header=False))
print("\n")

df.iloc[:, 1] = df.iloc[:, 1].rank() #rank experiment 1 from low to high
df.iloc[:, 2] = df.iloc[:, 2].rank() #rank experiment 2 from low to high
df.iloc[:, 3] = df.iloc[:, 3].rank() #rank experiment 3 from low to high
df.iloc[:, 4] = df.iloc[:, 4].rank() #rank experiment 4 from low to high
df_agg = df
df_agg['Avg'] = df_agg.iloc[:, 1:5].mean(axis=1) #find avg of rank and create column called "Avg"
df1_agg = df_agg.iloc[:,[0,1,2,3,4]] #df of rankings
print("***PART B***\n")
print(df1_agg.to_string(index=False, header=False)) #print rankings
print("\n")
df2_agg = df_agg.sort_values(by='Avg') #sort by avg ranking low to high
df2_agg = df2_agg.iloc[:,0]
print(df2_agg.to_string(index=False, header=False)) #print gene name sorted by avg ranking low to high
print("\n")

print("***PART C***\n")
df_agg.iloc[:, 1] = df_agg.iloc[:, 1].rank(method='first') #replace exp1 column with non-tied ranking
df_agg.iloc[:, 2] = df_agg.iloc[:, 2].rank(method='first') #replace exp2 column with non-tied ranking
df_agg.iloc[:, 3] = df_agg.iloc[:, 3].rank(method='first') #replace exp3 column with non-tied ranking
df_agg.iloc[:, 4] = df_agg.iloc[:, 4].rank(method='first') #replace exp4 column with non-tied ranking
df_first = df_agg #df with non-tied rankings
df_molecfunc = df_first.sort_values(by='GO: Molecular Function') #sort ranking for exp1 from low to high
molecfunc = list(df_molecfunc.index.values) #get index order of rankings for exp1
df_biopro = df_first.sort_values(by='GO: Biological Process') #sort ranking for exp2 from low to high
biopro = list(df_biopro.index.values) #get index order of rankings for exp2
df_cell = df_first.sort_values(by='GO: Cellular Component') #sort ranking for exp3 from low to high
cell = list(df_cell.index.values) #get index order of rankings for exp3
df_domain = df_first.sort_values(by='Domain') #sort ranking for exp4 from low to high
domain = list(df_domain.index.values) #get index order of rankings for exp4

def kendall(list1, list2):
    distance = 0
    for value1 in list1:
        for value2 in list2:
            index3 = list2.index(value1)
            index4 = list2.index(value2)
            index1 = list1.index(value1)
            index2 = list1.index(value2)
            if (index1 < index2 and index3 > index4) or (index1 > index2 and index3 < index4):
                distance += 1
            else:
                distance += 0
    return distance/2
#r1,r2
r12 = kendall(molecfunc, biopro)
#r1,r3
r13 = kendall(molecfunc, cell)
#r1,r4
r14 = kendall(molecfunc, domain)
r1 = r12 + r13 + r14 #sum of total rankings for exp1

#r2,r1
r21 = kendall(biopro, molecfunc)
#r2,r3
r23 = kendall(biopro, cell)
#r2,r4
r24 = kendall(biopro, domain)
r2 = r21 + r23 + r24 #sum of total rankings for exp2

#r3, r1
r31 = kendall(cell, molecfunc)
#r3, r2
r32 = kendall(cell, biopro)
#r3, r4
r34 = kendall(cell, domain)
r3 = r31 + r32 + r34 #sum of total rankings for exp3

#r4, r1
r41 = kendall(domain, molecfunc)
#r4, r2
r42 = kendall(domain, biopro)
#r4, r3
r43 = kendall(domain, cell)
r4 = r41 + r42 + r43 #sum of total rankings for exp4

mini = min(r1, r2, r3, r4) #find minimum of ranking sums
if mini == r1:
    print("GO: Molecular Function\n", df_molecfunc.iloc[:,0].to_string(index=False, header=False))
elif mini ==r2:
    print("GO: Biological Processs\n", df_biopro.iloc[:,0].to_string(index=False, header=False))
elif mini == r3:
    print("GO: Cellular Component\n", df_cell.iloc[:,0].to_string(index=False, header=False))
else:
    print("GO: Domain\n", df_domain.iloc[:,0].to_string(index=False, header=False))
```
